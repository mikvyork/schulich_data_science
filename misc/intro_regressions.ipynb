```python
import pandas as pd
import numpy as np
from statsmodels.api import OLS
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
bikes = pd.read_csv('https://raw.githubusercontent.com/delinai/schulich_ds1/main/Datasets/bikes_sharing.csv')
```

```python
bikes.head()
```

```python
bikes['datetime'] = pd.to_datetime(bikes['datetime'])
```

```python
bikes['weekday'] = bikes['datetime'].dt.weekday
```

```python
day_map = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
bikes['weekday'] = bikes['weekday'].map(day_map)
bikes
```

```python
bikes['hour'] = bikes['datetime'].dt.hour
```

```python
def get_time_of_day(x) : 
    if x in range (6, 10):
        return 'MorningCommute'
    elif x in range (10,15):
        return 'Midday'
    elif x in range(15,19): 
        return 'EveningCommute'
    else: 
        return 'Night'
```

```python
bikes['time_of_day'] = bikes['hour'].apply(get_time_of_day)
```

```python
bikes
```

```python
model_data = pd.get_dummies(bikes, columns=['season','weather','weekday','time_of_day','hour'], drop_first=True, dtype=int)
```

```python
model_data.columns
```

```python
Y = model_data['count']
X = model_data.drop(columns=['datetime','count', 'casual','registered'])
```

```python
model = OLS(Y,X).fit()
print(model.summary())
#p> means its statistically significant 
#sometimes the p value is a datapoint issue, not enough samples to draw conclusions with confidence
# but sometimes it truly means that there's no significant relationship between that variable and in that output 
```

```python
bikes[bikes['weekday']=='Thu']
# not a super meaningful day P value is likely real, in that any sort of pattern of coefficient is purely coicidential 
# not enough to determine statistical significance 
# it just means we cant really conclude that just because its a thursday that we are going to be renting 2 less bikes 
```

```python
y_pred = model.predict(X)
```

```python
sns.histplot(y_pred)
```

```python
res = y_pred - bikes['count']
sns.histplot(res)
```
